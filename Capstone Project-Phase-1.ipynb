```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime 
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("C:\\Users\\h473401\\State wise Data.csv")
```

```python
data.info()
```

```python
data['Population'] = data['Population'].str.replace(',', '').astype(int)
data['Area(Km2)'] = data['Area(Km2)'].str.replace(',', '').astype(int)
```

```python
data.info()
```

```python
# Plotting with rotated x-axis labels
plt.figure(figsize=(10, 6))
plt.bar(data['State'], data['Area(Km2)'], color='green')
plt.xlabel('State')
plt.ylabel('Area')
plt.title('Area vs. State')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# Plotting with rotated x-axis labels
plt.figure(figsize=(10, 6))
plt.bar(data['State'], data['Population'], color='blue')
plt.xlabel('State')
plt.ylabel('Population')
plt.title('Population vs. State')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# Plotting with rotated x-axis labels
plt.figure(figsize=(10, 6))
plt.bar(data['State'], data['Literacy'], color='red')
plt.xlabel('State')
plt.ylabel('Literacy')
plt.title('Literacy vs. State')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
